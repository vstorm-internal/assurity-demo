## Load image

```python
from pathlib import Path

image_dir = Path("./data/jpeg")
```

## Load model and processor

```python
from transformers import DonutProcessor, VisionEncoderDecoderModel

processor = DonutProcessor.from_pretrained("naver-clova-ix/donut-base-finetuned-cord-v2")
model = VisionEncoderDecoderModel.from_pretrained("naver-clova-ix/donut-base-finetuned-cord-v2")
```

## Prepare using processor

We prepare the image for the model using `DonutProcessor`.

# Inference in a random image
The model only accepts recipes; it needs to be fine-tuned for our purposes.

```python
import locale
locale.getpreferredencoding = lambda: "UTF-8"
import re

from PIL import Image
from tqdm import tqdm
import torch
```

```python
def donut_ocr(image_file):
    image = Image.open(image_file)
    pixel_values = processor(image, return_tensors="pt").pixel_values

    task_prompt = "<s_cord-v2>"
    decoder_input_ids = processor.tokenizer(task_prompt, add_special_tokens=False, return_tensors="pt")["input_ids"]

    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)

    outputs = model.generate(pixel_values.to(device),
                               decoder_input_ids=decoder_input_ids.to(device),
                               max_length=model.decoder.config.max_position_embeddings,
                               early_stopping=True,
                               pad_token_id=processor.tokenizer.pad_token_id,
                               eos_token_id=processor.tokenizer.eos_token_id,
                               use_cache=True,
                               num_beams=1,
                               bad_words_ids=[[processor.tokenizer.unk_token_id]],
                               return_dict_in_generate=True,
                               output_scores=True,)

    sequence = processor.batch_decode(outputs.sequences)[0]
    sequence = sequence.replace(processor.tokenizer.eos_token, "").replace(processor.tokenizer.pad_token, "")
    sequence = re.sub(r"<.*?>", "", sequence, count=1).strip()  # remove first task start token

    processor.token2json(sequence)
```

```python
donut_res = {}

for image_file in tqdm(image_dir.iterdir()):
    if not image_file.is_file():
        continue
    try:
        ocr_result = donut_ocr(image_file)
    except Exception as e:
        ocr_result = str(e)
    donut_res[str(image_file.stem)] = ocr_result
```
