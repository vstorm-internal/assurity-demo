```python
import os

# Replace 'YOUR_SERVICE_ACCOUNT_KEY.json' with the actual path to your service account key.
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/ocr-test-credentials.json"
```

```python
from google.cloud import vision


def detect_text(path):
    """Detects text in the file."""
    client = vision.ImageAnnotatorClient()
    with open(path, "rb") as image_file:
        content = image_file.read()
    image = vision.Image(content=content)
    response = client.text_detection(image=image)
    texts = response.text_annotations
    full_text = list()
    for text in texts:
        full_text.append(text.description)

    full_text = "\n".join(full_text)
    if response.error.message:
        raise Exception(
            f"{response.error.message}\nFor more info on error messages, check: https://cloud.google.com/apis/design/errors"
        )
    return full_text
```

```python
ocr_results = dict()
ocr_results["google_vision"] = dict()

for img in os.listdir("/content/data/"):
    res = detect_text("/content/data/" + img)
    ocr_results["google_vision"][img] = res
```
